```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import cv2
import os
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
file = '../../../Files/Datasets/preview/cat_0_5500.png'
k = 16
```

```python
img = cv2.imread(file)
```

```python
plt.imread(file).shape
```

```python
img=img/256
img.shape
```

```python
img.shape[0]
```

```python
img = np.reshape(img, (img.shape[1]*img.shape[0], 3))
```

```python
img.shape[1]
```

```python
centroids = np.zeros((k,img.shape[1]))
centroids.shape
```

```python
idx = np.random.randint(img.shape[0], size=img.shape[0])
idx
```

```python
img[idx[1:16],:]
```

```python
centroids = img[idx[1:16],:]
```

```python
initial_centroids = centroids
```

```python
initial_centroids[0]
```

```python
x_c=[]
y_c=[]
z_c=[]
for i in range(0, initial_centroids.shape[0]):
    x_c.append(initial_centroids[i][0])
    y_c.append(initial_centroids[i][1])
    z_c.append(initial_centroids[i][2])
```

```python
# Plot centroids
fig = plt.figure()
ax = Axes3D(fig)
plt.title('Initial Centroids')
ax.scatter(x_c, y_c, z_c, color='red')
plt.show()
```

```python
x=[]
y=[]
z=[]
for i in range(0, img.shape[0]):
    x.append(img[i][0])
    y.append(img[i][1])
    z.append(img[i][2])
```

```python

fig = plt.figure()
ax = Axes3D(fig)
plt.title('Image')
ax.scatter(x, y, z, color='green', label='image', marker='x', s=1)
ax.scatter(x_c, y_c, z_c, color='red', label='centroid', marker='o', s=100)
plt.legend(loc='upper left');
plt.show()
```

```python
# Kmeans using library
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)
clusters_knn = kmeans.fit_predict(img)
```

```python
x_0=[]
y_0=[]
z_0=[]
for i in range(0, img[clusters_knn==0].shape[0]):
    x_0.append(img[i][0])
    y_0.append(img[i][1])
    z_0.append(img[i][2])
    
x_1=[]
y_1=[]
z_1=[]
for i in range(0, img[clusters_knn==1].shape[0]):
    x_1.append(img[i][0])
    y_1.append(img[i][1])
    z_1.append(img[i][2])

x_2=[]
y_2=[]
z_2=[]
for i in range(0, img[clusters_knn==2].shape[0]):
    x_2.append(img[i][0])
    y_2.append(img[i][1])
    z_2.append(img[i][2])
```

```python
# Plot centroids
fig = plt.figure()
ax = Axes3D(fig)
plt.title('Initial Centroids')
ax.scatter(x_0, y_0, z_0, color='red', marker='x', s=100)
ax.scatter(x_1, y_1, z_1, color='green', marker='o', s=1)
ax.scatter(x_2, y_2, z_2, color='blue', marker='x', s=100)
plt.show()
```

```python
marker = ['x', 'o', 'x', 'o', 'x', 'o', 'o', 'x', 'o', 'x', 'o', 'x', 'o', 'x', 'o', 'x']
color = ['red', 'green', 'blue', 'yellow', 'pink', 'black', 'white', 'orange', 'red', 'green', 'blue', 'yellow', 'pink', 'black', 'white', 'orange']
ss = [100,1,10,1]
# Plot centroids
fig = plt.figure()
ax = Axes3D(fig)
plt.title('Initial Centroids')
for i in range(0,3):
    x_1=[]
    y_1=[]
    z_1=[]
    for j in range (0, img[clusters_knn==i].shape[0]):
        x_1.append(img[j][0])
        y_1.append(img[j][1])
        z_1.append(img[j][2])
    ax.scatter(x_1, y_1, z_1, color=color[i], marker=marker[i], s=ss[i])
    
plt.show()
```
